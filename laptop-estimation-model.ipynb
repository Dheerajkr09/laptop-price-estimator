```python
def get_processor(input):
    if input == "Intel Core i3" or input == "Intel Core i5" or input == "Intel Core i7":
        return input
    if input.split()[0] == "Intel":
        return "Other Intel Processor"
    if input.split()[0] == "AMD":
        return "Amd Processor"
    else:
        return "other Processor"
        
```

```python
df['Memory'] = df['Memory'].astype(str).str.replace(r'\.0', '', regex=True)  # .replace → .str.replace
df["Memory"] = df["Memory"].str.replace('GB', '', regex=False)
df["Memory"] = df["Memory"].str.replace('TB', '000', regex=False)

new = df["Memory"].str.split("+", n=1, expand=True)

df["first"] = new[0]
df["first"] = df["first"].str.strip()

df["second"] = new[1]

df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['first'] = df['first'].str.replace(r'\D', '', regex=True)   # ✅ regex=True

df["second"] = df["second"].fillna("0")                         # ✅ inplace warning fix

df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '', regex=True)  # ✅ regex=True

df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])

df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
                 'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
                 'Layer2Flash_Storage'], inplace=True)
```

```python
def getos(text):
    if text == "Windows 10 " or text == "Windows 7 " or text == "Windows 10 S":
         return "Windows"
    if text == "macOS" or text == "Mac OS X":
        return "MAC"
    if text == "Android " or text == "Linux":
        return "Android/Linux"
    else:
        return "others/No OS"
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns = "Price",axis = 1),np.log(df['Price']), test_size = 0.15,random_state = 2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,mean_absolute_error

```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

**Linear Regression**

```python
#apply OneHotEncoder
preprocessing1 = ColumnTransformer(transformers = [
    ("trf1",OneHotEncoder(sparse_output = False,drop = 'first',handle_unknown = "ignore"),[0,1,7,10,11]),
     ("trf2",StandardScaler(),[2,3,6,8,9])
    ],
    remainder = "passthrough")
   

model = LinearRegression()

pipe = Pipeline([
    ('preprocessing1',preprocessing1),
    ('model',model)
])

pipe.fit(x_train,y_train)

y_predict = pipe.predict(x_test)
print('R2 score',r2_score(y_test,y_predict))
print('MAE',mean_absolute_error(y_test,y_predict))
```
